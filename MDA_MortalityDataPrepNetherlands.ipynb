```python
df_MortalityNL_TEST = df_MortalityNL.melt(id_vars="Perioden",
                                         var_name = "COD",
                                         value_name = "deathcount")
```

```python
df_MortalityNL_TEST[['YEAR','quarters1', 'Kwartaal']]=df_MortalityNL_TEST.Perioden.str.split(" " ,expand=True)
print(df_MortalityNL_TEST)
```

```python
#make a year and quarters column
df_MortalityNL_TEST.drop(columns=["Perioden", "Kwartaal"])
```

```python
df_MortalityNL_order= df_MortalityNL_TEST [["YEAR", "quarters1", "COD", "deathcount"]]
df_MortalityNL_order.head(5)
```

```python
#change the values in quarters1
q = "SELECT year, case when quarters1 IN ('1e') then 'Q1' when quarters1  IN ('2e') then 'Q2' when quarters1 IN ('3e') then 'Q3' when quarters1 IN ('4e') then 'Q4' ELSE 'unknown' END as quarters, cod, deathcount FROM df_MortalityNL_order GROUP BY Year, quarters, cod"
newSQLtoPanda_df = sqldf(q, globals())
```

```python
newSQLtoPanda_df.head(10)
```

```python
newSQLtoPanda_df['COD'].value_counts()
```

```python
#I can only match Neoplasm, Diseases of the respiratory system and Diseases of the circulatory system 
t = "SELECT year, quarters, case when COD in ('Nieuwvormingen|Totaal nieuwvormingen') then 'Neoplasms' when COD in ('Ziekten van de ademhalingsorganen') then 'Diseases of the respiratory system' when COD in ('Ziekten van hart en vaatstelsel|Totaal ziekten van hart en vaatstelsel') then 'Diseases of the circulatory system' END as COD, deathcount from newSQLtoPanda_df"
newSQLtoPanda2_df = sqldf(t, globals())
```
