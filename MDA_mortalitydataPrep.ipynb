### Classificatie NL 'matchen' met BE  

---

NL: Kwaadaardige nv. van luchtpijp en long   
BE <-  neoplasme * % luchtpijp en long

----

NL: Nieuwvormingen|Kwaadaardige nieuwvorming van borst  Neoplasms  
BE <- neoplasme * % kwaadaardig borst

---

NL: Nieuwvormingen|Overige nieuwvormingen  
BE <- neoplasme * % overig

---

NL: Psychische stoornis. en zktn zenuwstel..  
BE <- Mental and behavioural disorders  + Diseases of the nervous system   

---

NL: Ziekten van hart en vaatstelsel|Totaal ziekten van hart en vaatstelsel  
BE <-  Diseases of the circulatory system x % hart en vaat..

---

NL: Ziekten van hart en vaatstelsel|Acuut hartinfarct  
BE <- Diseases of the circulatory system x % heart attack

---

NL: Ziekten van hart en vaatstelsel|Hersenvaatletsels  
BE <- Diseases of the circulatory system x % brain circulatory etc

---

NL: Ziekten van hart en vaatstelsel|Overige ziekten van hart en vaatstelsel  
BE <- Diseases of the circulatory system x % other

---

NL: Ziekten van de ademhalingsorganen  
BE <- Diseases of the respiratory system
